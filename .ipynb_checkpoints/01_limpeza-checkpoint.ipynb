# Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pathlib
from pathlib import Path
import os
import duckdb

pd.set_option('display.max_columns', None)
```

# Concatenação dos datasets

```python
BASE_PATH = Path("/home/matsa/Documents/Faculdade/SAD/tp01/raw_datasets/")
```

```python
def concat_datasets(datasets_path:pathlib.Path, output_filename):
    with duckdb.connect("db.db") as con:
        
        query = f"""
        COPY (
            SELECT * 
            FROM read_csv_auto('{datasets_path}/*.csv', union_by_name=True)
        )
        TO '{output_filename}' WITH (FORMAT CSV, HEADER TRUE);
        """

        con.execute(query)
        print(f"Files of {datasets_path} concatenated and saved to {output_filename}")

concat_datasets(BASE_PATH, "dados_violencia_mulheres_ses_concat_2.csv")
```

```python
df_concat = pd.read_csv("./dados_violencia_mulheres_ses_concat_2.csv", sep=",")

display(df_concat.head())
print(f"Shape: {df_concat.shape}")
```

#### Porcentagem de valores nulos de cada variável

```python
percent_nan_values = round((df_concat.isna().sum() * 100) / len(df_concat), 2)
percent_nan_values = percent_nan_values.sort_values(ascending=True)

plt.figure(figsize=(5,4))
sns.barplot(
    data= percent_nan_values
)
plt.xticks(ha="right", rotation=40)
plt.xlabel("Variables")
plt.ylabel("(%) NaN values")
plt.show()
```

# Limpeza

#### Removendo duplicatas

```python
duplicated_rows = df_concat[df_concat.duplicated(keep=False)]

# contagem da quantidade de duplicatas por por linhas completamente iguais
duplicated_count = duplicated_rows.groupby(
    list(df_concat.columns)
).size().reset_index(name="duplicates_count")

display(duplicated_count.head())
print("Duplicates count: ", duplicated_count.shape[0], "\n")
print(duplicated_count["duplicates_count"].value_counts())
```

```python
df_concat = df_concat.drop_duplicates()
```

#### Renomeando colunas

```python
df_concat = df_concat.rename({"dt_notific": "dt_notificacao", 
                  "dt_nasc": "dt_nascimento_vitima", 
                  "nu_idade_n": "idade_vitima", 
                  "cs_raca": "raca_vitima", 
                  "id_mn_resi": "municipio_residencia", 
                  "local_ocor": "local_ocorrencia", 
                  "out_vezes": "outras_vezes", 
                  "les_autop": "lesao_autoprovocada", 
                  "viol_fisic": "viol_fisica", 
                  "viol_psico": "viol_psicologica", 
                  "viol_sexu": "viol_sexual", 
                  "num_envolv": "num_envolvidos",
                  "orient_sex": "orient_sex_vitima", 
                  "ident_gen": "ident_gen_vitima"}, axis=1)
```

```python
df_concat.head()
```

#### Classificando tipos de violência

```python
def classify_violence(row):
    types = []

    if row["viol_fisica"] == "Sim":
        types.append("Física")
    if row["viol_psicologica"] == "Sim":
        types.append("Psicológica")
    if row["viol_sexual"] == "Sim":
        types.append("Sexual")

    return " | ".join(types) if types else "Não preenchido"
    
df_concat["tipo_violencia"] = df_concat.apply(classify_violence, axis=1)
```

#### Trabalhando valores nulos

```python
# --- Colunas numéricas --- #

# remove ',' da idade
df_concat["idade_vitima"] = df_concat["idade_vitima"].astype(str).str.split(",", expand=True)[0]
df_concat["idade_vitima"] = pd.to_numeric(df_concat["idade_vitima"], errors="coerce")

# preenchendo idades nulas com -1
df_concat["idade_vitima"] = df_concat["idade_vitima"].fillna(-1).astype("int16")

# convertendo datas para datetime
df_concat["dt_notificacao"] = pd.to_datetime(df_concat["dt_notificacao"], errors="coerce")
df_concat["dt_nascimento_vitima"] = pd.to_datetime(df_concat["dt_nascimento_vitima"], errors="coerce")

# calcula a idade da vítima baseada na data de notificação da ocorrência e a data de seu nascimento
df_concat["idade_calc"] = (df_concat["dt_notificacao"] - df_concat["dt_nascimento_vitima"]).dt.days // 365

# combina a idade da coluna original com a idade calculada. Se ainda tem registros nulos, preenche com -1
df_concat["idade_vitima"] = (
    df_concat["idade_calc"]
        .fillna(df_concat["idade_vitima"])
        .fillna(-1)
        .astype("int16")
)

# --- Colunas categóricas --- #

fillna_object_columns = ["raca_vitima", "municipio_residencia", "local_ocorrencia", "outras_vezes", "num_envolvidos", "lesao_autoprovocada", "autor_sexo", "orient_sex_vitima", "ident_gen_vitima"]

for col in fillna_object_columns:
    df_concat[col] = df_concat[col].fillna("Não preenchido")

display(df_concat.head())
print("Shape: ", df_concat.shape)
```

#### Convertendo tipos

```python
# alterando tipos
df_concat = df_concat.astype({
    "dt_nascimento_vitima": "datetime64[ns]",
    "municipio_residencia": "string", 
    "idade_vitima": "int16", 
    "raca_vitima": "string", 
    "orient_sex_vitima": "string", 
    "ident_gen_vitima": "string", 
    "dt_notificacao": "datetime64[ns]", 
    "local_ocorrencia": "string", 
    "num_envolvidos": "string", 
    "tipo_violencia": "string",
    "lesao_autoprovocada": "string", 
    "outras_vezes": "string"
})

# reordenando colunas
df_concat = df_concat[["dt_nascimento_vitima", "municipio_residencia", "idade_vitima", "raca_vitima", "orient_sex_vitima", "ident_gen_vitima", "dt_notificacao", "local_ocorrencia", "num_envolvidos", "tipo_violencia", "viol_fisica", "viol_sexual", "viol_psicologica", "lesao_autoprovocada", "outras_vezes", "autor_sexo"]]
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(
    data=df_concat["idade_vitima"]
)
plt.show()

print(df_concat["idade_vitima"].describe())

print("\nQuantidade de registros com idade -1: ", (df_concat["idade_vitima"] == -1).sum())

Q3 = df_concat["idade_vitima"].quantile(0.75)
Q1 = df_concat["idade_vitima"].quantile(0.25)
IQR = Q3 - Q1
upper_limit =  Q3 + (1.5 * IQR) 
print("\nQuantidade de registros de pessoas com idade acima do limite superior: ", (df_concat["idade_vitima"] > upper_limit).sum())
```

```python
# removendo valores com idade -1
df_concat = df_concat[df_concat["idade_vitima"] != -1]

# removendo municipio_residencia "Não Preenchido"
df_concat = df_concat[df_concat["municipio_residencia"] != "Não preenchido"]
```

```python
# devo remover idade 0 e violência autoprovocada?
df_concat[(df_concat["idade_vitima"] == 0) & (df_concat["lesao_autoprovocada"] == "Sim")]
```

#### Criando um id para o dataset

```python
df_concat.index = df_concat.index + 1
df_concat.index.name = "id"
```
